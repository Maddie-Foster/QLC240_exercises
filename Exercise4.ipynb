```python
import cartopy.crs as ccrs #imports the cartopy library for coordinate reference systems
import matplotlib.pyplot as plt #imports the matplotlib library for plotting

fig = plt.figure(figsize=(10, 5))#creates a new figure with a width of 10 inches and a height of 5 inches
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree()) #adds a subplot to the figure with a 1x1 grid, using the PlateCarree projection
ax.coastlines() #adds coastlines to the map
plt.title('Map of the Earth') #sets the title of the plot
plt.show() #displays the plot
```

In the following cell, add an arrow saying "We are here" that is pointing to Charlotte North Carolina


```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

#Add an arrow saying "We are here" that is pointing to Charlotte North Carolina
# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271
ax.annotate('We are here',
            xy=(charlotte_lon, charlotte_lat),  # Coordinates of the arrow tip
            xytext=(charlotte_lon - 10, charlotte_lat + 5),  # Coordinates of the text
            arrowprops=dict(facecolor='black', shrink=0.05), # Styling the arrow
            )


plt.title('Map of the Earth')
plt.show()
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Charlotte's coordinates
charlotte_lon = -80.8431
charlotte_lat = 35.2271

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.coastlines()

# Add the arrow and text

# Add major ocean labels (Atlantic, Pacific, etc.)
ax.text(-30, 20, 'Atlantic Ocean', transform=ccrs.PlateCarree(), fontsize=10)
ax.text(-150, 20, 'Pacific Ocean', transform=ccrs.PlateCarree(), fontsize=10)
ax.text(70, -30, 'Indian Ocean', transform=ccrs.PlateCarree(), fontsize=10)
ax.text(-100, -50, 'Southern Ocean', transform=ccrs.PlateCarree(), fontsize=10)
ax.text(-100, 70, 'Arctic Ocean', transform=ccrs.PlateCarree(), fontsize=10)


plt.title('Map of the Earth')
plt.show()
```
