```python
#Create a cell that imports the iris dataset from load_iris from sklearn.datasets

import sklearn.datasets
iris_df = sklearn.datasets.load_iris()

#Format iris_df into a pandas dataframe
iris_df = pd.DataFrame(data= np.c_[iris_df['data'], iris_df['target']],
                     columns= iris_df['feature_names'] + ['target'])
```
